**Question 1:**

Plot a histogram of the `sepal_width` from the `IrisDF` DataFrame below with 15 bins.

```python
import pandas as pd 
IrisDF = pd.read_csv("https://raw.githubusercontent.com/ThomasJewson/datasets/master/Iris.csv")
IrisDF
```

```python
#Solution
IrisDF.plot.hist(
    y="sepal_width",
    bins=15
)
```

**Question 2:**

*Below is some data of the average income per head in thousands of Euros. [1]. Output the following countries on a line plot: `Ireland`, `Greece`, `Luxembourg` and the `United Kingdom`.*

*Give the graph the following attributes:*

- Title of "Average income per head in thousands of Euros"

- Figure size of `(15,10)`



```python
IncomeDF = pd.read_csv("https://raw.githubusercontent.com/ThomasJewson/datasets/master/IncomePerHead2-1000EURO.csv")
IncomeDF
```

```python
#Solution
IncomeDF.plot.line(
    x="Year",
    y=["Ireland","Greece","Luxembourg","United Kingdom"],

    
    title="Average income per head in thousands of Euros",
    figsize=(15,10),
)
```

**Question 3**

Plot a bar chart with the `IncomeDF` DataFrame of the average income per head of per 1000 Euros of United States and France on the same plot.

Give the United States red bars and France blue bars. 

```python
#Solution
IncomeDF.plot.bar(
    x="Year",
    y=["United States","France"],
    color=["red","blue"]
)
```

**Question 4:**


*Below produces a pivot table from the `CheeseDF` we saw in the questions from Lesson 2.*

*Plot a pie chart of the `Units Sold` from the `CheesePiv` DataFrame*

```python
import numpy as np
CheeseDF = pd.read_csv("https://raw.githubusercontent.com/ThomasJewson/datasets/master/CheeseShop.csv")
CheesePiv=CheeseDF.pivot_table(index=["Cheese"],aggfunc=np.sum,values="Units Sold")
CheesePiv
```

```python
#Solution
CheesePiv.plot.pie(
    y="Units Sold"
)
```

**Question 5:**

*From the `IrisDF` plot the `petal_width` and the `petal_length` on the same histogram.*

*Give `petal_width` the colour `navy` and `petal_length` the colour `orange`.*

*Give the plot an appropriate title, grid lines and 15 bins.*

```python
#Solution
IrisDF.plot.hist(
    y=["petal_width","petal_length"],
    color = ["navy","orange"],
    title="Histogram of Petal widths and lengths",
    grid=True,
    bins=15
)
```
