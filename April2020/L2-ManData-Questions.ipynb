**Question 1:**

*Produce a pivot table from the `IrisDF` DataFrame below about `species`, which calculates the average values of `petal_length`,`petal_width`,`sepal_length` and `sepal_width`.*

*Output and save this pivot table to a new DataFrame*

*Hint: If you leave out the `aggfunc` argument, `pandas` will automatically find the average values.*

```python
import pandas as pd 
IrisDF = pd.read_csv("https://raw.githubusercontent.com/ThomasJewson/datasets/master/Iris.csv")
IrisDF
```

```python
# Solution
IrisPiv = IrisDF.pivot_table(index="species")
IrisPiv
```

**Question 2:**

*Calculate the average petal area of each species of flower and save this as a new column called `petal_area` in your DataFrame you made in question 1.*

```python
#Solution
IrisPiv["petal_area"] = IrisPiv["petal_length"] * IrisPiv["petal_width"]
IrisPiv
```

**Question 3:**

*Sort the `species` column in anti-alphabetical order*

```python
#Solution
IrisPiv.sort_values("species",ascending=False)
```

Sort the `sepal_width` column from smallest to largest

```python
#Solution
IrisPiv.sort_values("sepal_width")
```

**Question 4:**

*From the `PrisonDF` below, calculate the percentage full of each prison and save the data to a new column within the DataFrame called `Percentage Full`. Then, sort `Percentage Full` from largest to smallest.

```python
PrisonDF = pd.read_csv("https://raw.githubusercontent.com/ThomasJewson/datasets/master/PrisonsDataFull.csv")
PrisonDF
```

```python
#Solution
PrisonDF["Percentage Full"] = (PrisonDF["Population"] / PrisonDF["Capacity"])*100
PrisonDF.sort_values("Percentage Full",ascending=False)
```

**Question 5:**

*The `CheeseDF` DataFrame has sales data from a cheese shop. Produce a pivot table from `CheeseDF` which firsts pivots about the type of cheese then pivots about the quarter, producing a sum of just the `Units Sold` column.*

*Hint: You will need to import the `numpy` library and use the `numpy.sum` function.

```python
CheeseDF = pd.read_csv("https://raw.githubusercontent.com/ThomasJewson/datasets/master/CheeseShop.csv")
CheeseDF
```

```python
import numpy as np
CheeseDF.pivot_table(index=["Cheese","Quarter"],aggfunc=np.sum,values="Units Sold")
```
