###### What is a Library?

A library for Python is a collection of extra commands and functions you can use in Python, it can be thought as a plug-in that adds an extension to the functionality of Python. 

###### What does it mean to be open source?

When something is open source it means it can be downloaded and used for free. It also means you have access to the source code of the software, of which you can edit and modify. However, you cannot go on to sell this modified software as it must remain open source as well. 

There are many open source alternatives to proprietary software. For instance, *Microsoft Office* is proprietary therefore you must pay for it and you cannot access the source code. There are many open source alternatives to *Microsoft Office* the most popular being *LibreOffice* and *Apache OpenOffice*.

###### Library for Python example:

Below is a piece of code that imports our *pandas* Library. We must include this line of code if we want to use the functions from the library.


````python
import pandas # This line imports the pandas library
print("We have now successfully imported the pandas Library")
````

We can now use all the functions of the library now that it has been imported. However, to call a function we need to use the following syntax ```pandas.FUNCTION```.

In the code below we have an example that produces a series. Don't worry if you don't understand what is going on, the key thing is to grasp the way we call functions from the pandas library

````python
st_yearsreleased_series = pandas.Series([1997,1980,1983,1999,2002,2005])
#See how we have had to use pandas.Series() to call the function, not just Series()
st_yearsreleased_series
````

To save us from having to write down `pandas.FUNCTION` everytime we can give the pandas library a nickname when we import it. Notice how we can now use `pd.Function` as we have given pandas the nickname of `pd`. 

````python
import pandas as pd
MySeries = pd.Series(["I","like","Python"])
MySeries
````

###### What is a DataFrame?

When we import tabular data into pandas, it stores it as in an object know as a DataFrame. An object is a collection of variables. These variables include infomation on the data, headers and indexes.

To output the DataFrame we just need to call the objects name. 

Below is some code which 'hard-codes' data into the `starwar_df` object. You do not need to worry how this code works.

````python
starwars_data = [
    ["A New Hope",1977,"8/10"],
    ["The Empire Strikes Back",1980,"9/10"],
    ["Return of the Jedi",1983,"8/10"],
    ["The Phantom Menace",1999,"6/10"],
    ["Attack fo the Clones",2002,"6/10"],
    ["Revenge of the Sith",2005,"7/10"],
]
#Above is the hard-coded data


starwars_df = pd.DataFrame(starwars_data,columns=[
        "Name",
        "Year Released",
        "Rating"
    ]
)
````

To ouput the data from `starwars_df` all we need to do is call its name. See below.

````python
starwars_df
````

**Artificial Intelligence**

Articial Intelligence (AI) is the ability of a computer program to think and to learn. Despite what you might think AI is not about walking talking robots that act and behave like humans. In fact, AI is just a method to automate certain tasks through the use of a program. One of the key aspects about AI is its ability to predict future outcomes based off data on previous outcomes.

For example, AI is being employed to predicted when credit or debit cards are used for fraud [1]. This means, everytime you use your card a probability is predicted by the AI, of the chance that your card is being used for fraud. The AI tracks your location and spending habits to determine this value. If the probability is high enough your card will get rejected and your money will be safe from fraudsters.

AI can be broken down into different techniques of learning. For instance, a subsection of AI is machine learning and then a subsection of machine learning is deep learning / neural networks.

**Big data and its applications**

Before the widespread use of computers, data was hard to come by as it required lots of people to collect, manipulate and analyse it - which was expensive and slow. Now when we connect to a website every single thing we do is stored as data, from the time you look at a single picture to the amount you scroll - even whether you click an advertisement or not - everything is saved as data. We are in a time where the collection of data is growing at an unimaginable rate, where more than 2.5 million terabytes of data is produced every single day.

The term 'Big data' refers not just to the total amount of data produced but also to the new algorithmic techniques which extract infomation from this dataset. Some of these new data analysis techniques are AI, machine learning and deep learning, as mentioned above. 

Big data is changing the world and in partcular it is changing the world of healthcare. For example, every year influenza (or the flu) spreads through many countries, leading to considerable loss of life. With the use of big data the Google Flu Trends team stared work on predicting when and where flu epidemics would occur. The Google Flu Trends team analysed search engine queries relevant to the flu, and from this the team predicted where and when outbreaks might occur. However, they were famously wrong and over-predicted the 2009 epidemic by 50% and the 2012–13 epidemic by 140%. Despite these problems using internet search related data shows potential in healthcare. [2][3]

With such a large amount of data we can predict and find correlations that we previously havent been able to do. This is why learning how to do statistics and data analysis on Python is so important. As Mathematician Clive Humby said "Data is the new oil.".
